```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np

import brewer2mpl
# brewer2mpl.get_map args: set name set type number of colors
bmap = brewer2mpl.get_map('Paired', 'qualitative', 10)
new_colors = bmap.mpl_colors
# new_colors = ['seashell', 'honeydew', 'lightcyan']


labels = ['CCW', 'ECW', 'VOCW']

colors = ['lightcoral', 'orange', 'lightgreen', 'cornflowerblue', 'darkviolet', 'seashell', 'honeydew', 'lightcyan']
markers = ['*', 'd', 's', '^', 'o']
font_size = 50
legend_size = 45
markersize = 20
plt.figure(figsize=(17, 12))  
plt.rcParams['font.family'] = 'Times New Roman'  # 设置字体
x_label = 'Traffic scenario'
#################################################################### #################################################################

# plt.figure(figsize=(16, 12))
plt.xlabel(x_label, fontsize=font_size)
plt.ylabel('F1 Score', fontsize=font_size)

x_vals = [1, 2, 3, 4, 5]
new_x_index = ['Scenario 1', 'Scenario 2', 'Scenario 3', 'Scenario 4', 'Scenario 5']

ccw_precision =     [0.743902439, 0.633928571, 0.880597015, 0.588235294, 0.624365482]
ecw_precision =        [0.776595745, 0.690909091, 0.90369898, 0.647058824, 0.744565217]
vccw_precision =     [0.8625, 0.728070175, 0.928162746, 0.710526316, 0.762886598]

ccw_recall =     [0.37195122, 0.29338843, 0.553877551, 0.298507463, 0.361764706]
ecw_recall =        [0.445121951, 0.314049587, 0.578367347, 0.328358209, 0.402941176]
vccw_recall =     [0.420731707, 0.342975207, 0.590137429, 0.402985075, 0.435294118]

ccw = []
ecw = []
vccw = []
for i in range(len(ccw_precision)):
    ccw.append(2 * ccw_precision[i] * ccw_recall[i] / ( ccw_precision[i] + ccw_recall[i] ))
    ecw.append(2 * ecw_precision[i] * ecw_recall[i] / ( ecw_precision[i] + ecw_recall[i] ))
    vccw.append(2 * vccw_precision[i] * vccw_recall[i] / ( vccw_precision[i] + vccw_recall[i] ))


diff = 1
x = np.arange(1, 6, 1)

width = diff * 0.12
small_width = diff * 0.025
label_text_size = 28


plt.bar(x - width - 2 * small_width, ccw, width, hatch="/", color=new_colors[0], edgecolor=None, label=labels[0])
plt.bar(x - width + 4 * small_width, ecw, width, hatch="-", color=new_colors[2], edgecolor=None, label=labels[1])
plt.bar(x + 5.5 * small_width, vccw, width, hatch="x", color=new_colors[4], edgecolor=None, label=labels[2])




plt.grid()

plt.ylim(0, 0.9)  
plt.xticks(x_vals, new_x_index, fontsize=font_size)
# plt.yticks(y_vals, new_y_index, fontsize=font_size)

# 刻度值字体大小设置
plt.tick_params(labelsize=40)
plt.legend(fontsize = 35, loc=0, ncol=3)
# plt.show()
#################################################################### #################################################################

plt.savefig("Fig6-5c-different-scenarios.pdf", bbox_inches='tight')
plt.show()
```
