```python

import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np

import brewer2mpl
# brewer2mpl.get_map args: set name set type number of colors
bmap = brewer2mpl.get_map('Paired', 'qualitative', 10)
colors = bmap.mpl_colors
# new_colors = ['seashell', 'honeydew', 'lightcyan']


labels = ['CCW', 'ECW', 'VCCW']

markers = ['*', 'd', 's', '^', 'o']
font_size = 50
legend_size = 45
markersize = 20
plt.figure(figsize=(17, 12))  
plt.rcParams['font.family'] = 'Times New Roman'  # 设置字体
x_label = 'Packet loss rate (%)'
#################################################################### #################################################################

# plt.figure(figsize=(16, 12))
plt.xlabel(x_label, fontsize=font_size)
plt.ylabel('F1 Score', fontsize=font_size)

x_vals = [1, 2, 3, 4, 5]
new_x_index = ['0.0', '1.5', '3.0', '4.5', '6.0']

ccw_precision =     [0.884476534, 0.890350877, 0.88239159, 0.880689655, 0.885861561]
ecw_precision =        [0.903412377, 0.90596745, 0.902035623, 0.89973262, 0.908119658]
vccw_precision =     [0.925501433, 0.933688573, 0.92893401, 0.925925926, 0.940213523]

ccw_recall =     [0.568885449, 0.566135458, 0.555417701, 0.550905953, 0.531360424]
ecw_recall =        [0.604489164, 0.598804781, 0.58643507, 0.580672994, 0.563162544]
vccw_recall =     [0.618773946, 0.622336227, 0.6, 0.599315068, 0.57736014]

ccw = []
ecw = []
vccw = []
for i in range(len(ccw_precision)):
    ccw.append(2 * ccw_precision[i] * ccw_recall[i] / ( ccw_precision[i] + ccw_recall[i] ))
    ecw.append(2 * ecw_precision[i] * ecw_recall[i] / ( ecw_precision[i] + ecw_recall[i] ))
    vccw.append(2 * vccw_precision[i] * vccw_recall[i] / ( vccw_precision[i] + vccw_recall[i] ))

x = np.arange(1, 6, 1)

plt.plot(x_vals, ccw, linestyle=":", marker=markers[0], markersize=markersize, color=colors[0], label=labels[0], linewidth=5.0, markerfacecolor='none', markeredgewidth = 5.0, markeredgecolor = colors[1])
plt.plot(x_vals, ecw, linestyle=":", marker=markers[1], markersize=markersize, color=colors[2], label=labels[1], linewidth=5.0, markerfacecolor='none', markeredgewidth = 5.0, markeredgecolor = colors[3])
plt.plot(x_vals, vccw, linestyle=":", marker=markers[2], markersize=markersize, color=colors[4], label=labels[2], linewidth=5.0, markerfacecolor='none', markeredgewidth = 5.0, markeredgecolor = colors[5])


plt.grid()

# plt.ylim(0, 1.1)  
plt.xticks(x_vals, new_x_index, fontsize=font_size)
# plt.yticks(y_vals, new_y_index, fontsize=font_size)

# 刻度值字体大小设置
plt.tick_params(labelsize=font_size)
plt.legend(fontsize = 35, loc=0, ncol=3)
# plt.show()
#################################################################### #################################################################

plt.savefig("Fig6-7c-different-packet-loss-rate.pdf", bbox_inches='tight')
plt.show()
```
